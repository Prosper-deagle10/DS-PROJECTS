```python
# Import the required libraries for data analysis process
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
age = np.random.randint(17,27,(300,1))
data = pd.DataFrame(age)

# Adding a random 'Dept' column
dept = np.random.choice(['Computer Science', 'Maths/Stat', 'Economics', 'Finance Banking', 'Others'], size=len(data))
data['Department'] = dept

# Adding a random 'Level' column
level = np.random.choice(['300', '400'], size=len(data))
data['Level'] = level

# Adding a random 'University' column
uni = np.random.choice(['Uniport', 'Ignatius', 'RSU', 'Capt Elechi Amadi'], size=len(data))
data['University'] = uni

# Display the resulting DataFrame
print(data)
```

```python
# Rename the unnamed column to 'NewName'
data.rename(columns={data.columns[0]: 'Age'}, inplace=True)

# Display the resulting DataFrame
print(data)
```

```python
# Let;s find the How many Universities in the data
data['University'].value_counts().sort_values(ascending=True)\
.plot(kind='bar',figsize=(8,6),color=['#457b9d','#ef233c','#06d6a0'])
plt.title("Universities in Port Harcourt")
plt.xlabel('University')
plt.ylabel("Count of the values")
plt.show()
```

```python
# Let;s find the How many Departments in the data
data['Department'].value_counts().sort_values(ascending=True)\
.plot(kind='bar',figsize=(8,6),color=['#457b9d','#ef233c','#06d6a0'])
plt.title("Universities in Port Harcourt")
plt.xlabel('Department')
plt.ylabel("Count of the values")
plt.show()
```

```python
# Let;s find the How many levels in the data
data['Level'].value_counts().sort_values(ascending=True)\
.plot(kind='bar',figsize=(8,6),color=['#457b9d','#ef233c'])
plt.title("Universities in Port Harcourt")
plt.xlabel('Level')
plt.ylabel("Count of the values")
plt.show()
```

```python
# Let;s find the How many Age in the data
data['Age'].value_counts().sort_values(ascending=True)\
.plot(kind='pie',figsize=(8,6))
plt.title("Universities in Port Harcourt")
plt.xlabel('Age')
plt.ylabel("Count of the values")
plt.show()
```
